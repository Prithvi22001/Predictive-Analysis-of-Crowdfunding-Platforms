```
# Import Libraries
import numpy as np
import pandas as pd
import glob
from google.colab import drive
drive.mount('/content/drive')
```

```
# Load the dataset
df = pd.read_csv(r'/content/drive/My Drive/Predictive Analysis of Crowdfunding Platforms/4.Rewards Dataset (Initial)/Rewards Dataset.csv', error_bad_lines=False, engine='python')
print(df.shape)
df.head()
```

```
# Check for duplicate rows
duplicateDf = df[df.duplicated(['name', 'url'])] 
print(duplicateDf)
```

```
# Remove duplicate rows
df.drop_duplicates(subset=['name', 'url'], keep='first', inplace=True)
df.tail()
```

```
# Reset Index
df.reset_index(drop=True, inplace=True)
df.tail()
```

```
# Drop columns which are not useful
df.drop(['project id', 'name', 'url', 'funded percentage', 'funded date'], axis=1, inplace=True)
print(df.shape)
df.head()
```

```
# Analysis of the Project State
df['status'].value_counts()
```

```
# Consider only successful and failed projects
df = df[(df['status'] == 'successful') | (df['status'] == 'failed')]
df.tail()
```

```
# Reset Index
df.reset_index(drop=True, inplace=True)
df.tail()
```

```
# Label Encoding of Project Status
df['status'] = df['status'].replace({'successful': 1, 'failed': 0})
df.head()
```

```
# Check for Null Values
df.isnull().sum()
```

```
# Drop Null Values
df.dropna(inplace=True)
df.shape
```

```
# Reset Index
df.reset_index(drop=True, inplace=True)
df.tail()
```

```
# Analysis of Project Location
df['location'] = df['location'].str.split(', ').str[1]
df.head()
```

```
with np.printoptions(threshold=np.inf):
  print(df['location'].unique())
```

```
df['location'].value_counts()
```

```
# Drop Null Values
df.dropna(inplace=True)
df.shape
```

```
# Reset Index
df.reset_index(drop=True, inplace=True)
df.tail()
```

```
# Replace all states in US with United States
united_states_array = set()
for i in range(len(df)):
  # print(df.loc[i, 'location'])
  if len(df.loc[i, 'location']) == 2:
    united_states_array.add(df.loc[i, 'location'])
print(united_states_array)
```

```
united_states_array = list(united_states_array)
df["location"].replace(united_states_array, "United States", inplace=True)
df.head()
```

```
df['location'].value_counts()
```

```
# Round up Goal Amount
df['goal'] = df['goal'].astype(float).round().astype(int)
df.head()
```

```
# Round up Pledged Amount
df['pledged'] = df['pledged'].astype(float).round().astype(int)
df.head()
```

```
# Round up Project Duration
df['duration'] = df['duration'].astype(float).round().astype(int)
df.head()
```

```
# Analysis of Category Column
df['category'].value_counts()
```

```
df['category'] = df['category'].replace({'Film &amp; Video': 'Film & Video'})
df['category'].value_counts()
```

```
# Analysis of Sub Category Column
df['subcategory'].value_counts()
```

```
df['subcategory'] = df['subcategory'].replace({'Film &amp; Video': 'Film & Video', 'Country &amp; Folk': 'Country & Folk', 'Board &amp; Card Games': 'Board & Card Games'})
df['subcategory'].value_counts()
```

```
# Rename columns for consistency
df.rename(columns={'category':'main_category', 'subcategory':'sub_category', 'location': 'country', 'reward levels': 'reward_levels'}, inplace=True)
df.head()
```

### **Reward Levels**

```
df.dtypes
```

```
df['reward_levels'] = df['reward_levels'].str.replace(',', '')
df['reward_levels'] = df['reward_levels'].str[1:]
df.head()
```

```
df['reward_levels'] = [list(map(int, rewards.split('$'))) for rewards in df['reward_levels']]
df.head()
```

```
import statistics
min_reward = []
max_reward = []
mean_reward = []
median_reward = []
for i in range(len(df)):
  arr = df.loc[i, 'reward_levels']
  min_reward.append(min(arr))
  max_reward.append(max(arr))
  mean_reward.append(round(statistics.mean(arr)))
  median_reward.append(round(statistics.median(arr)))
```

```
df['min_reward'] = min_reward
df['max_reward'] = max_reward
df['mean_reward'] = mean_reward
df['median_reward'] = median_reward
df.head()
```
