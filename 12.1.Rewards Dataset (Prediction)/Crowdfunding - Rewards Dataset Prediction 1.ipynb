```
# Import Libraries
import numpy as np
import pandas as pd
import glob
from google.colab import drive
drive.mount('/content/drive')
```

```
# Load the dataset
df = pd.read_csv(r'/content/drive/My Drive/Predictive Analysis of Crowdfunding Platforms/11.Rewards Dataset (Final)/Rewards Dataset Cleaned Final.csv')
print(df.shape)
df.head()
```

```
df['country'] = df['country'] // 10
```

### Pledged Amount Prediction

```
X = df.drop(['pledged','reward_levels','status'], axis=1)
X.head()
```

```
y = df['pledged']
y.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

LINEAR REGRESSION

```
from sklearn.linear_model import LinearRegression
reg = LinearRegression(normalize=True).fit(X_train, y_train)
reg.score(X_test,y_test)
```

```
y_pred = reg.predict(X_test)
```

```
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```
c = pd.DataFrame(reg.coef_,X_test.columns, columns=['Coeff'])
c
```

```
# Cross Validation Score
from sklearn.model_selection import cross_val_score
reg_eval = cross_val_score(estimator = reg, X = X_train, y = y_train, cv = 10)
reg_eval.mean()
```

RANDOM FOREST

```
df = pd.read_csv(r'/content/drive/My Drive/Predictive Analysis of Crowdfunding Platforms/11.Rewards Dataset (Final)/Rewards Dataset Cleaned Final.csv')
df['country'] = df['country'] // 10
```

```
X = df.drop(['pledged','reward_levels','status'], axis=1)
X.head()
```

```
y = df['pledged']
y.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```
from sklearn.ensemble import RandomForestRegressor
for i in range(100, 1001, 100):
  regressor = RandomForestRegressor(n_estimators = i, random_state = 42)
  regressor.fit(X_train, y_train)
  print(regressor.score(X_test,y_test))
```

```
y_pred = regressor.predict(X_test)
```

```
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df.head(50)
```

```
regressor.score(X_test,y_test)
```

```
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```
# Cross Validation Score
from sklearn.model_selection import cross_val_score
regressor_eval = cross_val_score(estimator = regressor, X = X_train, y = y_train, cv = 10)
regressor_eval.mean()
```

DECISION TREE

```
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(random_state=0)
dtr.fit(X_train,y_train)
dtr.score(X_test,y_test)
```

```
y_pred = dtr.predict(X_test)
```

```
df=pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
df.head(50)
```

```
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```
# Cross Validation Score
from sklearn.model_selection import cross_val_score
dtr_eval = cross_val_score(estimator = dtr, X = X_train, y = y_train, cv = 10)
dtr_eval.mean()
```
