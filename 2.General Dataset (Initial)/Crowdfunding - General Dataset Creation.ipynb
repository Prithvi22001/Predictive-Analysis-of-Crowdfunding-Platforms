```
# Import Libraries
import numpy as np
import pandas as pd
import glob
from google.colab import drive
drive.mount('/content/drive')
```

```
# There are 57 separate datasets with same number of columns. We combine all of them into 1 single dataframe
path = r'/content/drive/My Drive/Predictive Analysis of Crowdfunding Platforms/1.General Dataset (Initial Files)'
all_files = glob.glob(path + "/*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
frame = pd.concat(li, axis=0, ignore_index=True)
frame.head()
```

```
# Column Information
frame.info()
```

```
# Column Statistics
frame.describe()
```

```
# Store the newly obtained dataframe in the form of CSV File. This will be the final general dataset, which we will be used for pre processing
frame.to_csv('General Dataset.csv', index=False)
```
