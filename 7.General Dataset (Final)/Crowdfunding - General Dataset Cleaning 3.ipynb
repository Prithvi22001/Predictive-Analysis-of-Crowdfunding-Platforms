```
from datetime import datetime
from currency_converter import CurrencyConverter
c = CurrencyConverter(fallback_on_missing_rate=True)
```

```
goal_amount_usd = []
pledged_amount_usd = []
for i in range(len(df)):
  date = datetime.strptime(df.loc[i, 'launched_at'], "%Y-%m-%d %H:%M:%S")
  year = date.year
  month = date.month
  day = date.day
  goal_amount_usd.append(round(c.convert(df.loc[i, 'goal_amount'], df.loc[i, 'currency'], 'USD', date=datetime(year, month, day))))
  pledged_amount_usd.append(round(c.convert(df.loc[i, 'pledged_amount'], df.loc[i, 'currency'], 'USD', date=datetime(year, month, day))))
```

```
print(len(goal_amount_usd))
print(len(pledged_amount_usd))
```

```
df['goal_amount_usd'] = goal_amount_usd
df['pledged_amount_usd'] = pledged_amount_usd
df.head()
```

```
df.to_csv('Cleaned 2.csv', index=False)
```
