```
# Import Libraries
import numpy as np
import pandas as pd
import glob
from google.colab import drive
drive.mount('/content/drive')
```

```
# Load the dataset
df = pd.read_csv(r'/content/drive/My Drive/Predictive Analysis of Crowdfunding Platforms/3.Content Dataset (Initial)/Content Dataset.csv')
print(df.shape)
df.head()
```

```
# Check for duplicate rows
duplicateDf = df[df.duplicated(['Name'])] 
print(duplicateDf)
```

```
# Remove duplicate rows
df.drop_duplicates(subset=['Name'], keep='first', inplace=True)
df.tail()
```

```
# Reset Index
df.reset_index(drop=True, inplace=True)
df.tail()
```

```
# Drop columns which are not useful
df.drop(['Id', 'Name', 'Url', 'Creator', 'Updates', 'Comments', 'Start', 'End', 'Latitude', 'Longitude', 'Start Timestamp (UTC)', 'End Timestamp (UTC)', 'Creator Bio'], axis=1, inplace=True)
print(df.shape)
df.head()
```

```
# Rename columns for consistency
df.rename(columns={'State':'Status', 'Top Category':'Main Category', 'Category':'Sub Category', 'Duration in Days':'Duration', 'Creator Website':'Has Website', 'Creator - # Projects Created':'Projects Created', 'Creator - # Projects Backed':'Projects Backed', '# Videos':'Videos', '# Images':'Images', '# Words (Description)':'Description Words', '# Words (Risks and Challenges)':'Risks and Challenges Words', '# FAQs':'FAQ'}, inplace=True)
df.head()
```

```
df.dtypes
```

```
# Analysis of the Project State
df['Status'].value_counts()
```

```
# Label Encoding of Project Status
df['Status'] = df['Status'].replace({'successful': 1, 'failed': 0})
df.head()
```

```
# Check for Null Values
df.isnull().sum()
```

```
# Replace Null Values with 0
df.fillna(value=0, inplace=True)
df.isnull().sum()
```

```
# Analysis of the Project State
df['Currency'].value_counts()
```

```
# Analysis of Main Category
df['Main Category'].value_counts()
```

```
# Analysis of Sub Category
df['Sub Category'].value_counts()
```

```
# Analysis of Location
df['Location'].value_counts()
```

```
# All Locations are of United States. Hence this column can be dropped.
df.drop(['Location'], axis=1, inplace=True)
print(df.shape)
df.head()
```

```
# Analysis of Facebook Connected
df['Facebook Connected'].value_counts()
```

```
# Label Encoding of Facebook Connected
df['Facebook Connected'] = df['Facebook Connected'].replace({'Yes': 1, 'No': 0})
df.head()
```

```
# Round up Facebook Friends
df['Facebook Friends'] = df['Facebook Friends'].astype(float).round().astype(int)
df.head()
```

```
# Analysis of Has Video
df['Has Video'].value_counts()
```

```
# Label Encoding of Has Video
df['Has Video'] = df['Has Video'].replace({'Yes': 1, 'No': 0})
df.head()
```

```
# Label Encoding of Has Website
df.loc[df['Has Website'] != 0, 'Has Website'] = 1
df.head()
```

```
# Round up Projects Backed
df['Projects Backed'] = df['Projects Backed'].astype(float).round().astype(int)
df.head()
```

```
# Round up Videos
df['Videos'] = df['Videos'].astype(float).round().astype(int)
df.head()
```

```
# Round up Risks and Challenges Words
df['Risks and Challenges Words'] = df['Risks and Challenges Words'].astype(float).round().astype(int)
df.head()
```
